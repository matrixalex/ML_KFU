```python
import pandas as pd
```

```python
df = pd.read_csv('iris_data.csv')
```

```python
df.head(10)
```

```python
%matplotlib inline
import seaborn as sns
```

```python
df['class'].unique()
i = 0
for cls in df['class'].unique():
    df.replace(cls, i, inplace=True)
    i += 1
```

```python
sns.pairplot(df, hue='class')
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
x = df.loc[:, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['class']
print(len(x), len(y))
```

```python
from sklearn.model_selection import train_test_split

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2, shuffle=True)
```

```python
print(len(y_train), len(x_train))
```

```python
snn = KNeighborsClassifier(n_neighbors=30)
```

```python
snn.fit(x_train, y_train)
```

```python
y_pred = snn.predict(x_test)
print(sum(y_pred - y_test))
```
